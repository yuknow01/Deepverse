```python
import torch

# 파일 불러오기
checkpoint = torch.load("best_gru_earlyfusion.pt")

# 1. 어떤 이름표(Key)들이 저장되어 있나?
print("보관함 항목:", checkpoint.keys()) 
# 출력: dict_keys(['epoch', 'model_state', 'optimizer_state', 'best_val', 'F_in', 'F_out'])

# 2. 모델의 가중치 이름들만 훑어보기
weights = checkpoint['model_state']
print("\n첫 3개 레이어의 가중치 이름과 크기:")
for i, (name, param) in enumerate(weights.items()):
    if i < 3:
        print(f"이름: {name:30} | 크기: {param.shape}")

# 3. 실제 특정 레이어의 '값'을 살짝 구경하기
sample_weight = weights['head.weight']
print("\n마지막 출력층 가중치 일부:\n", sample_weight[0, :5]) # 첫 번째 행의 앞 5개 숫자만 출력
```
