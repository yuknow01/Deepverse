# Data shape
Deepverse DAtaset

[Root / Scenario]<p>
- dataset.params['dataset_folder'] = "scenarios" <p>
- dataset.params['scenario'] = "DT31" <p>
- dataset.scenario_path = "scenarios/DT31"<p>
- dataset.scenario = ScenarioManage(...)<p>

[Coverage]<p>
- dataset.params['scenes'] = [1~100] -> here can change "scenarios/DT31/param/config.m"<p>
- dataset.params['basestations'] = [1]<p>


[Sensor/Data Toggles]
- dataset.params['camera']    = True
- dataset.params['camera_id'] = ['unit1_cam1']
- dataset.params['lidar']     = True
- dataset.params['lidar_id']  = ['unit1_lidar1']
- dataset.params['position']  = True

# Under important setting

[Enabled Modalities (Sub-dataset Handles)]
- dataset.camera_dataset   = CameraDataset(...) # Camera data
- dataset.lidar_dataset    = LiDARDataset(...)  # LiDAR data
- dataset.mobility_dataset = MobilityDataset(...)  # mobility/pose/position data
- dataset.comm_dataset     = CommunicationDataset(...) # wireless communication (OFDM channel) data
- dataset.radar_dataset    = RadarDataset(...)  # radar (FMCW) data


[Communication Setting: dataset.params['comm']]
- enable = True
- BS antenna: shape=[16,1], rotation=[0,0,-45.04], spacing=0.5, FoV=[360,180]
- UE antenna: shape=[1,1],  rotation=[0,0,0],      spacing=0.5, FoV=[360,180]
- OFDM :
- - bandwidth = 0.05
  - subcarriers = 512
  - selected_subcarriers = [0..7]
- activate_RX_filter = 0
- generate_OFDM_channels = 1
- num_paths = 25
- enable_Doppler = 1

[Radar Setting: dataset.params['radar']]
- TX antenna: shape=[1,1],  rotation=[0,0,-45.04], spacing=0.5, FoV=[180,180]
- RX antenna: shape=[16,1], rotation=[0,0,-45.04], spacing=0.5, FoV=[180,180]
- FMCW:
- - chirp_slope = 8.014e12
  - Fs = 6.2e6
  - n_samples_per_chirp = 256
  - n_chirps = 256
- num_paths = 50000
