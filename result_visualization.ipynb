# GRU scene 100
<br> trainable params : 2692096

```python
import re
import matplotlib.pyplot as plt

log_path = "gru_scene100.txt"
exp_name = "GRU Scene 100"

pat = re.compile(
    r"\[(\d+)/(\d+)\]\s*"
    r"train loss=([0-9.eE+-]+),\s*nmse\(dB\)=([0-9.eE+-]+)\s*\|\s*"
    r"val loss=([0-9.eE+-]+),\s*nmse\(dB\)=([0-9.eE+-]+)"
)

epochs = []
train_losses, val_losses = [], []
train_nmse_db, val_nmse_db = [], []
total_epochs = None

with open(log_path, "r", encoding="utf-8") as f:
    for line in f:
        m = pat.search(line)
        if not m:
            continue

        ep = int(m.group(1))
        total_epochs = int(m.group(2))
        tr_l = float(m.group(3))
        tr_n = float(m.group(4))
        va_l = float(m.group(5))
        va_n = float(m.group(6))

        epochs.append(ep)
        train_losses.append(tr_l)
        val_losses.append(va_l)
        train_nmse_db.append(tr_n)
        val_nmse_db.append(va_n)

if not epochs:
    raise RuntimeError("로그에서 학습 라인을 하나도 못 찾았어. 포맷이 다른지 확인해줘.")

# ---- Plot 1: Loss ----
plt.figure()
plt.plot(epochs, train_losses, label="train loss")
plt.plot(epochs, val_losses, label="val loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title(f"{exp_name} - Train vs Val Loss")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# ---- Plot 2: NMSE(dB) ----
plt.figure()
plt.plot(epochs, train_nmse_db, label="train nmse (dB)")
plt.plot(epochs, val_nmse_db, label="val nmse (dB)")
plt.xlabel("Epoch")
plt.ylabel("NMSE (dB)")
plt.title(f"{exp_name} - Train vs Val NMSE (dB)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
# ---- Best (lowest) Val NMSE(dB) ----
best_val_nmse = min(val_nmse_db)
best_idx = val_nmse_db.index(best_val_nmse)
best_epoch = epochs[best_idx]

print(f"[BEST] val nmse(dB) = {best_val_nmse:.4f} at epoch {best_epoch}/{total_epochs}")
```
